```
df = pd.read_csv('/content/drive/My Drive/ventas ferris.txt', sep='\t')
df.head()
```

```
#Le damos formato de fecha (mes)
df['Month'] = pd.to_datetime(df['Month'])
```

```
#Al mes se pone como indice
df.set_index('Month', inplace=True)
df.head()
```

```
#Para ver el rango de fechas
print(df.index.min())
print(df.index.max())
```

```
df.shape
```

```
#Cantidad de registros por a√±os
print(len(df['1964']))
print(len(df['1972']))
```

```
#ver estadisticas basicas
df.describe()
```

```
#Se grafica
df.plot()
```

```
#Calculo de la media por mes
meses = df.resample('M').mean()
meses
```

```
#Comparacion de las ventas mensuales de los a√±os 1971 y 1972
plt.plot(meses['1970'].values,color='r')
plt.plot(meses['1971'].values,color='b')
plt.plot(meses['1972'].values,color='g')
```

```
#Comparo 2 periodos de tiempo
verano1970 = df['1969-01-01':'1970-12-01']
plt.plot(verano1970.values)
verano1971 = df['1971-01-01':'1972-12-01']
plt.plot(verano1971.values)
```
